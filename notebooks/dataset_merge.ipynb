```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
weather_df = pd.read_csv('../data/curated/weather_daily.csv')
```

```python
print(f"Number of instances: {len(weather_df)}")
print(f"Number of features : {len(weather_df.columns)}")
```

```python
weather_df.tail(5)
```

```python
# remove the specified column from the DataFrame
weather_df = weather_df.drop(weather_df.columns[0], axis=1)
print(weather_df.head())
```

```python
tlc_taxi_data = pd.read_csv('../data/curated/processed_taxi_data.csv')
```

```python
tlc_taxi_data.tail(5)
```

```python
# columns to merge
left_column = 'date'
right_column = 'DATE'
drop_column = right_column

merged_df = pd.merge(tlc_taxi_data, weather_df, left_on=left_column, right_on=right_column)
final_df = merged_df.drop(columns=drop_column)

final_df
```

```python
null_summary = {col: final_df[col].isna().sum() for col in final_df.columns}
for column, count in null_summary.items():
    print(f"'{column}': \t {count} null values.")
```

```python
import os
processed_dir = "../data/curated"
os.makedirs(processed_dir, exist_ok=True)

final_df.to_csv(os.path.join(processed_dir, "final_dataset.csv"))
```
