```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/curated/final_dataset.csv')
```

```python
df
```

```python
df = df.drop(columns=['Unnamed: 0'])
print(df.head())
```

```python
df.describe()
```

```python
feature_list = df.columns.tolist()

print(feature_list)
```

```python
# feature columns for predictors
feature_columns = [
    'PULocationID', 'total_with_tip', 'trip_distance', 'fare_amount', 'tip_amount',
    'TEMP', 'DEWP', 'WDSP', 'MXSPD', 'GUST', 'MAX', 'MIN', 'PRCP'
]
predictors = df.loc[:, feature_columns]
print(predictors.head())
```

```python
response = df[['earnings_rate']]
```

```python
from sklearn.model_selection import train_test_split

# split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(predictors, response, test_size=0.4, shuffle=False)
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
# split the test set into validation and test sets
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test,
    test_size=0.5, shuffle=True, random_state=42)
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=100, random_state=42)
regressor.fit(X_train, y_train)

# feature importances and indices of top 8 features
importances = regressor.feature_importances_
top_indices = np.argsort(importances)[-8:]

selected_columns = X_train.columns[top_indices].tolist()
print(selected_columns)
```

```python
feature_importances = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': importances
})
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

print(feature_importances)

plt.figure(figsize=(10, 6))
plt.barh(feature_importances['Feature'], feature_importances['Importance'])
plt.xlabel('Importance')
plt.title('Feature Importances from RandomForestRegressor')
plt.gca().invert_yaxis()
plt.show()
```

As can be seen from this table and graph, PRCP (Precipitation amount) isn't very relevant and so can be removed.

While MAX (Max temperature), TEMP (Avg Temp) and MIN (Min Temp) can be arguably removed as it can be seen as contributing very little to the model's predictive power, these features do have significant meaning in the context of weather in relation to Taxi services. Therefore, retaining these features might prove beneficial although low in terms of importance.
